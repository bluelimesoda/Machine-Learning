```python
dataset = pd.read_csv('/Users/katekim/Desktop/PythonMLWorkspace(LightWeight)/ScikitLearn/LinearRegressionData.csv')
```

```python
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:,-1].values 
```

```python
X,y
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(X,y)
```

```python
y_pred = reg.predict(X)
y_pred
```

```python
plt.scatter(X,y, color='pink')
plt.plot(X,y_pred, color = 'orange')
plt.title('Test by Hour')
plt.xlabel('Hours')
plt.ylabel('Score')
plt.show()
```
